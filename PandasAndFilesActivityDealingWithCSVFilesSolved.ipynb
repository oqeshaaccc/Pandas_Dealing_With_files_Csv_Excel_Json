```python
#import pandas
import pandas as pd
```

```python
#to read data from CSV files.
#read csv --- pd.read_csv("")
csvfile = pd.read_csv("cities.csv", header= None)
csvfile.tail(5)
```

```python
csvfile.columns=["LatD","LatM","LatS","NS","LonD","LonM","LonS","EW","City","State"]
csvfile
```

```python
csvfile = csvfile.drop(index=0, axis=0)#used to delete the first row that was the old header
csvfile
```

```python
csvfile = csvfile.reset_index() #used to reset the index
csvfile
```

```python
csvfile = csvfile.drop(columns=['index'])#used to drop the column index that appeared after the reset of the indexing
csvfile
```

```python
csvfile["LatD"]=csvfile["LatD"].astype(int) #change the datatype of each column
csvfile["LatM"]=csvfile["LatM"].astype(int)
csvfile["LatS"]=csvfile["LatS"].astype(int)
csvfile["NS"]=csvfile["NS"].astype(object)
csvfile["LonD"]=csvfile["LonD"].astype(int)
csvfile["LonM"]=csvfile["LonM"].astype(int)
csvfile["LonS"]=csvfile["LonS"].astype(int)
csvfile["EW"]=csvfile["EW"].astype(object)
csvfile["City"]=csvfile["City"].astype(object)
csvfile["State"]=csvfile["State"].astype(object)
```

```python
#new_column = pd.Series([1, 2, 3, 4, 5]) #use it to add a new column with specific value to each row,
#but numbers of values has to equal number of rows

csvfile['NewColumn'] = 0 #this adds a new column with value 0 to all raws
csvfile
```
