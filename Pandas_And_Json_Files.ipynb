```python
#import pandas
import pandas as pd
```

```python
#read json file
jsonfile = pd.read_json("Jay.json")
```

```python
jsonfile.head()
```

```python
#add a column to the dataframe
jsonfile["NewKey"] = 0
jsonfile.head()
```

```python
#to add new row in dataset(append)
new_row = {'Duration': 55, 'Pulse': 121, 'Maxpulse': 151, 'Calories': 423,'NewKey': 100}
jsonfile = jsonfile.append(new_row, ignore_index=True)
```

```python
jsonfile.tail()
```

```python
#perform Describe on dataframe
jsonfile.describe()
```

```python
#perform info on dataframe
jsonfile.info()
```

```python
#save and export the dataframe in a new file
jsonfile.to_json("Jay2.json")
```
